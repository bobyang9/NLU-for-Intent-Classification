```python
import re
```

```python
intent = ""
lines = []
with open("train.txt") as file:
    for line in file:
        if(len(line) == 0):
            continue
        if(line[0] == '#'):
            intent = line[10:-1]
        elif(line[0] == '-'):
            without_paren = re.sub("\(.*?\)", "", line[2:-1])
            without_bracket = re.sub("[\[|\]]", "", without_paren)
            if(len(without_bracket) != 0 and without_bracket[-1] in '.?!'):
                without_bracket = without_bracket[:-1]
            without_bracket = without_bracket.lower()
            new_line = intent + ", " + without_bracket + '\n'
            lines.append(new_line)
            
```

```python
print(lines)
```

```python
file1 = open("snips_train_actual.csv","w")

file1.writelines(lines)

file1.close()
```
